<center> <h4 style="background-color:blue; color:white"><br> Sigmoid Activation Function <br></h3></center>


**Used in `output layer` when problem is `Binary Classification Problem`** <br>
Most popularly used activation Function
<h3> $$σ(z) = \frac{1}{1+e^{-z}}$$ </h3>

For any input -> z sigmoid will be given by above formula
> `Input Range`: (-∞ to +∞) i.e. it can take any real values<br>
> `Output Range`: (0 to 1)    

* Sigmoid function `asymptotes to zero at negative infinity` and `asymptotes to one at positive infinity` but there are intermediate values all in between
* Sigmoid Activation Function is mostly used at the Output layer where `we need Probability of classes as output`

**`Derivative-Sigmoid Activation Function`** <br>
<h3> $$\frac {dσ}{dz} = σ(z)(1-σ(z))$$ </h3>
It's derivative will give very small values see bellow

---
* A non-linear function so can capture more complex patterns
* Output values are bounded so don’t get too large
* Can suffer from **`Vanishing gradient`**

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''
    ''' 
    Sigmoid function is considered not good since near boundaries network doesn't learn 
    quickly.This is because gradient is almost zero near boundaries
    '''
    return 1/(1+np.exp(-x))

x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Sigmoid Activation Function')
plt.show()
```

```python
# https://stackoverflow.com/questions/49977063/role-derivative-of-sigmoid-function-in-neural-networks
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def derivative(x, step):
    return (sigmoid(x+step) - sigmoid(x)) / step

def other_derivative(x):
    y = [(1/(1 + np.exp(-i))) * (1 - (1 /(1 + np.exp(-i)))) for i in x]
    return y

x = np.linspace(-10, 10, 1000)

y1 = sigmoid(x)
y2 = derivative(x,y1)
y3 = other_derivative(x)

plt.plot(x, y1, label='sigmoid')
plt.plot(x, y2, label='derivative of sigmoid')
plt.plot(x, y3, label='other derivative')
plt.legend(loc='upper left')
plt.show()
```

<center> <h4 style="background-color:blue; color:white"><br> Tanh Activation Function || Hyperbolic Tangent Activation Function <br></h4></center>


Hyperbolic Tangent Activation Function is a `scaled and shifted sigmoid function`

<h3>$$tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$$</h3>
or
<h3>$$tanh(z) = \frac{2}{1 + e^{-2z})}-1$$</h3>
or
<h3>$$tanh(z) = 2σ(2z)-1$$</h3>
While comparing sigmoid and tanh we can see we have factor of 2 and -1 

> `Input Range`: (-∞ to +∞) i.e. it can take any real values<br>
> `Output Range`: (-1 to 1)

**`Derivative-tanh Activation Function`** <br>
<h3>$$\frac {dtanh}{dz} = 1 - tanh^2(z)$$</h3>

`NOTE:`<br>
Both `Sigmoid and Tanh` are great choices because they are 
* `differentiable at any point` 
* `monotonic in nature`
* `smooth` 

Problems such as `saturation would occur` due to 
* either high or 
* low input values 

to function ending up in `asymptotic plateaus of function`

Since curve is almost flat at these points, `derivatives are almost zero` therefore the `weight updates become very slow or even halts`
* If `gradients are close to zero` then `weight updates becomes slow due to small step sizes down the hill during gradient descent`
* this doesn’t enable us to create complex chain of function that we’ll need to capture relations in data

---
* A non-linear function so can capture more complex patterns
* Output values are bounded so don’t get too large
* Can suffer from **Vanishing gradient**

```python
def tanh(x):
    ''' It returns value (1-exp(-2x))/(1+exp(-2x))
    Tanh also suffers from gradient problem near boundaries just as Sigmoid activation 
    function does
    '''
    return np.tanh(x)

x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Tanh Activation Function')
plt.show()
```

<center> <h4 style="background-color:blue; color:white"><br> ReLu Activation Function (Rectified Linear Unit) <br></h4></center>

RELU is less computational expensive than other non linear activation functions and mostly used
* As negative portion of input space translating to zero activation ReLU layers can end up dying which can cause training to stop

* `RELU returns 0 if z (input) is less than 0 or negative values`
* `RELU returns z if z (input) is greater than 0`

For +ve values it will act as linear function

<h3>$$ReLu(z) = max(0,z)$$</h3>

> `Input Range`: (-∞ to +∞) i.e. it can take any real values<br>
> `Output Range`: (0 to ∞)

**`NOTE`**<br>
* Derivate of negative values comes up as 0
* Derivate of positive values comes up as 1
* Not differentiable at point 0
    * For implementation purpose we assume Derivative of ReLu at 0 as 0
    
**`Problem`**
* As derivative for some neurons becomes 0 so `Weight` and `Biases` of such neurons do not update

`Leaky ReLU` comes in rescue then

---
* A non-linear function so can capture more complex patterns
* Values can get very large
* As it does not allow for negative values, certain patterns may not be captured
* Gradient can go towards 0 so weights are not updated: “dying ReLU problem”

```python
def RELU(x):
    ''' 
    It returns zero if input is less than zero 
    otherwise it returns given input
    '''
    x1=[]
    
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)

    return x1

x = np.linspace(-10, 10)
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('ReLu Activation Function')
plt.show()
```

<center> <h4 style="background-color:blue; color:white"><br> Leaky ReLU <br></h4></center>

This modified version of ReLue allow small negative values when input is less than zero.

When ReLU is saturated and not active Leaky-ReLU rectifier allows a small non-zero gradient

<h3>$$LeakyReLu(z) = max(0.01z,z)$$</h3>

**`Derivative:`** <br>
* 0.01 for all values less then or equll to 0
* 1 for all values greater then 0

So during back propogation all weight and bias will be updated

```python
def dlrelu(x,alpha=0.01):
    dx = np.ones_like(x)
    dx[x < 0] = alpha
    return dx

x = np.linspace(-10,10)
plt.plot(dlrelu(x,alpha=0.01))
plt.axis('tight')
plt.title('Leaky ReLu Activation Function')
plt.show()
```

<center> <h4 style="background-color:blue; color:white"><br> Softmax <br></h4></center>

**Used in `output layer` when problem is `Multiclass Classification Problem`** <br>

* Turns logits, numeric output of last linear layer of a multi-class classification neural network into probabilities
* It calculates relative probabilityies
* Sigmoid only return probabilityies

[Diff b/w Sigmoid and Softmax](https://stats.stackexchange.com/questions/233658/softmax-vs-sigmoid-function-in-logistic-classifier)

<h3>$$softmax(z_i) = \frac {exp(z_i)}{\sum_jexp(z_j)}$$</h3>
$$z_i ==> values-from-neurons-in-output-layer$$
$$exp(z_i) => acts-as-non-linear-function$$
$${\sum_jexp(z_j)} => normalize-values-convert-to-probabilities$$

If number of classes will be two this formula will work as Sigmoid Function
* can say sigmoid is a variante of softmax function

```python
def softmax(x):
    ''' 
    For computing softmax values for each sets of scores in x 
    '''
    return np.exp(x) / np.sum(np.exp(x),axis=0)

x = np.linspace(-10, 10)
plt.plot(x,softmax(x))
plt.axis('tight')
plt.title('Softmax Activation Function')
plt.show()
```

## Steps involved in Gradient Descent Process
1. Take values of parameter`w`(weight) and `b`(bias), calculate Cost function
2. Take step to steepest downill direction i.e. direction where error will reduce, by updating parameters w and b
3. Repeat previous steps until minima is achieved

## Criteria to stop Gradient Descent Process
1. Error is not updating
2. Number of iterations exausted

* [How Gradient Descent works](https://builtin.com/data-science/gradient-descent)
* [implement-a-gradient-descent-in-python](https://www.geeksforgeeks.org/how-to-implement-a-gradient-descent-in-python-to-find-a-local-minimum/)

In gradient descent method of optimization, a hypothesis function
$$hθ(x)$$
is fitted to a data set
$$(x(i),y(i)) --> (i=1,2,⋯,m) $$
by minimizing an associated cost function
$$J(θ)$$ in terms of parameters θ=00,θ1,⋯ <br> 
Cost function describes how closely hypothesis fits data for a given choice of θ

`Example:`<br> 
one might wish to fit a given data set to a straight line,
<h4>$$hθ(x)=θ0+θ1x$$</h4>
An appropriate cost function might be sum of squared difference between data and hypothesis:
<h3>$$J(θ)=\frac {1}{2m}\sum^m_i[hθ(x^i)−y^i]^2$$</h3>
The gradient descent method starts with a set of initial parameter values of θ (say, θ0=0,θ1=0), and then follows an iterative procedure, changing the values of θj so that J(θ) decreases:
<h4>$$θj→θj−α \frac {∂}{∂θj}J(θ)$$</h4>
To simplify things, consider fitting a data set to a straight line through the origin: $$hθ(x)=θ1x$$
In this one-dimensional problem, we can plot a simple graph for J(θ1) and follow the iterative procedure which trys to converge on its minimum.

```python
import warnings
warnings.filterwarnings('ignore')
### one-parameter plot 

#data to fit
m = 20
theta1_true = 0.5
x = np.linspace(-1,1,m)
y = theta1_true * x

# The plot: LHS is the data, RHS will be the cost function.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6.15))
ax[0].scatter(x, y, marker='x', s=40, color='k')

def cost_func(theta1):
    """The cost function, J(theta1) describing the goodness of fit."""
    theta1 = np.atleast_2d(np.asarray(theta1))
    return np.average((y-hypothesis(x, theta1))**2, axis=1)/2

def hypothesis(x, theta1):
    """Our "hypothesis function", a straight line through the origin."""
    return theta1*x

# First construct a grid of theta1 parameter pairs and their corresponding
# cost function values.
theta1_grid = np.linspace(-0.2,1,50)
J_grid = cost_func(theta1_grid[:,np.newaxis])

# The cost function as a function of its single parameter, theta1.
ax[1].plot(theta1_grid, J_grid, 'k')

# Take N steps with learning rate alpha down the steepest gradient,
# starting at theta1 = 0.
N = 5
alpha = 1
theta1 = [0]
J = [cost_func(theta1[0])[0]]
for j in range(N-1):
    last_theta1 = theta1[-1]
    this_theta1 = last_theta1 - alpha / m * np.sum((hypothesis(x, last_theta1) - y) * x)
    theta1.append(this_theta1)
    J.append(cost_func(this_theta1))

# Annotate the cost function plot with coloured points indicating the
# parameters chosen and red arrows indicating the steps down the gradient.
# Also plot the fit function on the LHS data plot in a matching colour.
colors = ['b', 'g', 'm', 'c', 'orange']
ax[0].plot(x, hypothesis(x, theta1[0]), color=colors[0], lw=2,
           label=r'$\theta_1 = {:.3f}$'.format(theta1[0]))
for j in range(1,N):
    ax[1].annotate('', xy=(theta1[j], J[j]), xytext=(theta1[j-1], J[j-1]),
                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                   va='center', ha='center')
    ax[0].plot(x, hypothesis(x, theta1[j]), color=colors[j], lw=2,
               label=r'$\theta_1 = {:.3f}$'.format(theta1[j]))

# Labels, titles and a legend.
ax[1].scatter(theta1,J, c=colors, s=40, lw=0)
ax[1].set_xlim(-0.2,1)
ax[1].set_xlabel(r'$\theta_1$')
ax[1].set_ylabel(r'$J(\theta_1)$')
ax[1].set_title('Cost function')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
ax[0].set_title('Data and fit')
ax[0].legend(loc='upper left', fontsize='small')

plt.tight_layout()
plt.show()
```

```python
## two-dimensional plotting

#data to fit
m = 20
theta0_true = 2
theta1_true = 0.5
x = np.linspace(-1,1,m)
y = theta0_true + theta1_true * x

# The plot: LHS is the data, RHS will be the cost function.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,6.15))
ax[0].scatter(x, y, marker='x', s=40, color='k')

def cost_func(theta0, theta1):
    """The cost function, J(theta0, theta1) describing the goodness of fit."""
    theta0 = np.atleast_3d(np.asarray(theta0))
    theta1 = np.atleast_3d(np.asarray(theta1))
    return np.average((y-hypothesis(x, theta0, theta1))**2, axis=2)/2

def hypothesis(x, theta0, theta1):
    """Our "hypothesis function", a straight line."""
    return theta0 + theta1*x

# First construct a grid of (theta0, theta1) parameter pairs and their
# corresponding cost function values.
theta0_grid = np.linspace(-1,4,101)
theta1_grid = np.linspace(-5,5,101)
J_grid = cost_func(theta0_grid[np.newaxis,:,np.newaxis],
                   theta1_grid[:,np.newaxis,np.newaxis])

# A labeled contour plot for the RHS cost function
X, Y = np.meshgrid(theta0_grid, theta1_grid)
contours = ax[1].contour(X, Y, J_grid, 30)
ax[1].clabel(contours)
# The target parameter values indicated on the cost function contour plot
ax[1].scatter([theta0_true]*2,[theta1_true]*2,s=[50,10], color=['k','w'])

# Take N steps with learning rate alpha down the steepest gradient,
# starting at (theta0, theta1) = (0, 0).
N = 5
alpha = 0.7
theta = [np.array((0,0))]
J = [cost_func(*theta[0])[0]]
for j in range(N-1):
    last_theta = theta[-1]
    this_theta = np.empty((2,))
    this_theta[0] = last_theta[0] - alpha / m * np.sum(
                                    (hypothesis(x, *last_theta) - y))
    this_theta[1] = last_theta[1] - alpha / m * np.sum(
                                    (hypothesis(x, *last_theta) - y) * x)
    theta.append(this_theta)
    J.append(cost_func(*this_theta))


# Annotate the cost function plot with coloured points indicating the
# parameters chosen and red arrows indicating the steps down the gradient.
# Also plot the fit function on the LHS data plot in a matching colour.
colors = ['b', 'g', 'm', 'c', 'orange']
ax[0].plot(x, hypothesis(x, *theta[0]), color=colors[0], lw=2,
           label=r'$\theta_0 = {:.3f}, \theta_1 = {:.3f}$'.format(*theta[0]))
for j in range(1,N):
    ax[1].annotate('', xy=theta[j], xytext=theta[j-1],
                   arrowprops={'arrowstyle': '->', 'color': 'r', 'lw': 1},
                   va='center', ha='center')
    ax[0].plot(x, hypothesis(x, *theta[j]), color=colors[j], lw=2,
           label=r'$\theta_0 = {:.3f}, \theta_1 = {:.3f}$'.format(*theta[j]))
ax[1].scatter(*zip(*theta), c=colors, s=40, lw=0)

# Labels, titles and a legend.
ax[1].set_xlabel(r'$\theta_0$')
ax[1].set_ylabel(r'$\theta_1$')
ax[1].set_title('Cost function')
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$y$')
ax[0].set_title('Data and fit')
axbox = ax[0].get_position()
# Position the legend by hand so that it doesn't cover up any of the lines.
ax[0].legend(loc=(axbox.x0+0.5*axbox.width, axbox.y0+0.1*axbox.height),fontsize='small')

plt.show()
```

#### Update equation for Gradient Descent Algorithm
<h3>$$\theta_i = \theta_i - \alpha * \frac {dJ}{d\theta_i}$$</h3>
theta is parameter we wish to update
$$\theta ==> Parameter$$
Partial derivative which tells us the rate of change of error or the cost function with respect to the parameter theta and alpha here is the learning rate
$$\frac {dJ}{d\theta_i} ==> Gradient-or-Slope$$

$J$ ==> represent-cost-function <br>
$\alpha$ ==> learning rate


<center> <h4 style="background-color:blue; color:white" ><br> Batch Gradient Descent <br></h4></center>

1. If we use Entire training set say(`m`) to find out `J` i.e. cost function, then it is **`Batch Gradient Descent`** <br>
We take entire training set(m), 
* perform forward propagation and calculate cost function.
* Then we `update parameters` using rate of change of this cost function with respect to parameters

`Epoch` is when entire training set is passed through model `forward propagation` and `backward propagation` is performed and parameters are updated

In `Batch gradient descent` as we are using entire training set, parameters will be updated only once per epoch 


<center> <h4 style="background-color:blue; color:white" ><br> Stochastic Gradient Descent(SGD) <br></h4></center>


2. If we use `single observation` to calculate Cost Function, then it is **`Stochastic Gradient Descent(SGD)`** <br>
We pass a single observation at a time, calculate cost and update parameters
Each time parameter is passed is called `Iteration` <br>
There will be M iteration per epocs:
* M => number of observation in dataset

<center> <h4 style="background-color:blue; color:white" ><br> Mini-Batch Gradient Descent <br></h4></center>


3. If we use subset of dataset which must be `1 < x < m` to calculate Cost Function, then it is **`Mini-Batch Gradient Descent`**
* x => Batch size
    * it must be more then 1 and less then m
    * m => entire training set

### Comparison: Number of Observations used for Updation 

|Batch Gradient Descent| Stochastic Gradient Descent(SGD)| Mini-Batch Gradient Descent|
|-----|-----|-----|
Take Entire dataset, calculate cost function then update parameters like w,b |Take single observation from dataset at a time, calculate cost function then update parameters (iteration)  | Take subset(batch) dataset, calculate cost function then update parameters like w,b 
Cost function reduces smoothly | Lot of variation in cost function | Smoother cost function as compared to SGD
Computation cost is very high | Computation time is more | Computation time less then SGD

**1. It gets stuck at `local minima`** <br>
Stochastic Gradient Descent(SGD) with momentum
<h3>$$\theta_i = \theta_i - \alpha * V_t$$</h3>
with this we can solve problem of getting stuck at Local Minima


<center> <h3 style="background-color:orange; color:white" ><br> Implementing Stochastic Gradient Descent(SGD) with_Momentum from Scratch <br></h3></center>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating input array
X=np.array([[1,0,1,0], [1,0,1,1], [0,1,0,1]])
# converting input in matrix form
X = X.T
print ('\n Input:')
print(X)
# shape of input array
print('\n Shape of Input:', X.shape)

# creating output array
y=np.array([[1],[1],[0]])
# output in matrix form
y = y.T
print ('\n Actual Output:')
print(y)
# shape of output array
print('\n Shape of Output:', y.shape,'\n')
```

```python
# defining  Sigmoid Function
def sigmoid (x):
    return 1/(1 + np.exp(-x))
```

```python
# defining hyperparameters of model
lr=0.01 # learning rate
inputlayer_neurons = X.shape[0] # number of features in data set
hiddenlayer_neurons = 3 # number of hidden layers neurons
output_neurons = 1 # number of neurons at output layer

epochs = 10000 # number of epochs

# initializing weight
w_ih = np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
w_ho = np.random.uniform(size=(hiddenlayer_neurons,output_neurons))

# defining momentum parameters
beta = 0.9
v_ho = 0
v_ih = 0

error_momentum = []
for i in range(epochs):
    ##### Forward Propogation ##### 
    
    # hidden layer activations
    hidden_layer_input=np.dot(w_ih.T,X)
    hiddenlayer_activations = sigmoid(hidden_layer_input)    
    # calculating the output
    output_layer_input=np.dot(w_ho.T,hiddenlayer_activations)
    output = sigmoid(output_layer_input)
    
    
    ##### Backward Propagation #####
    # calculating error
    error = np.square(y-output)/2
    error_wrt_output = -(y-output)
    output_wrt_Z2 = np.multiply(output,(1-output))
    Z2_wrt_who = hiddenlayer_activations
    # rate of change of error w.r.t weight between output and hidden layer
    error_wrt_who = np.dot(Z2_wrt_who,(error_wrt_output*output_wrt_Z2).T)
    Z2_wrt_h1 = w_ho
    h1_wrt_Z1 = np.multiply(hiddenlayer_activations,(1-hiddenlayer_activations))
    Z1_wrt_wih = X
    # rate of change of error w.r.t weights between input and hidden layer
    error_wrt_wih = np.dot(Z1_wrt_wih,(h1_wrt_Z1*np.dot(Z2_wrt_h1,(error_wrt_output*output_wrt_Z2))).T)

    # weighted gradient
    v_ho = beta * v_ho + (1-beta)*error_wrt_who
    w_ho = w_ho - lr * v_ho
    v_ih = beta * v_ih + (1-beta)*error_wrt_wih
    w_ih = w_ih - lr * v_ih
    
    # appending the error of each epoch
    error_momentum.append(np.average(error))
    
    
# visualizing error after each epoch
plt.figure(figsize=(15,5))
plt.plot(np.arange(1,epochs+1), np.array(error_momentum))
plt.show()
```

**2. `Same learning rate throughout training process`** <br>
Learning rate should update as per Cost Function <br>

<h3>$$\eta = \sum^{t-1}_{i=1}{[\frac{dJ}{d\theta}]^2}$$</h3>
It is possible if we will use `RMSProp`
<h3>$$μ_t = \beta μ_{t-1}+ (1 - \beta) [\frac {dJ}{d \theta}]^2$$</h3>

`RMSProp` changes learning rate as the trainig progresses

# Loss Function for Regression Problem:
1. `MAE`
$$MAE = \frac{1}{N} \sum^N_{i=1}|y_i - \hat y_i|$$
2. `MSE`
$$MSE = \frac{1}{N} \sum^N_{i=1}(y_i - \hat y_i)^2$$
3. `RMSE`
$$MSE = \sqrt {\frac{1}{N} \sum^N_{i=1}(y_i - \hat y_i)^2}$$

# Loss Function for Classification Problem:
[Log Loss Function Derivation](https://www.youtube.com/watch?v=hWLdFMccpTY) <br>
1. For **`Binary Classification Problem`** || in keras we have 
    1. **`Binary Cross Entropy or Log Loss`** => negative average of the log of corrected prediceted probabilities
        * We will have `Actual class` and there `Predicted Probability`
        * `Predicted Probability` column have all prob. that data point belongs to class-1
        * `Corrected Probability` column will have probability 1-Predicted Probability, where we have class-0 in Actual class subtract there Predicted Probability with 1 to get actual probability of class-1
        * As `Predicted Probability` column have probability of class being 1 not 0
        * Take log of `Corrected Probability` and multply by negative sign

2. For **`Multi-Class Classificatin Problem`** use loss function as
    1. **`Categorical Cross Entropy`**

## 3. (Scaling)-Bringing all variables in range 0 to 1

1. **`Min-Max Scaling:`** <br>
Re-scales a feature or observation value with distribution value between 0 and 1
<h3>$$X_{new} = \frac {X_i - X_{min}}{X_{max} - X_{min}}$$</h3>

2. **`Standardization:`** <br>
Very effective technique which re-scales a feature value so that it has distribution with 0 mean value and variance equals to 1
<h3>$$X_{new} = \frac {X_i-X_{mean}}{Standard Deviation}$$</h3>

```python
#before scaling
data['ApplicantIncome'].head()
```

```python
# applying min-max scaling manually
### bringing variables in range 0 to 1

data['ApplicantIncome'] = (data['ApplicantIncome'] - data['ApplicantIncome'].min()) / (data['ApplicantIncome'].max() - data['ApplicantIncome'].min())
```

```python
import keras
from keras import Input, Model
from keras.layers import InputLayer,Dense,concatenate
```

```python
### Defining architecture of model using Keras Functional API #####

#### Defining Model architecture using Keras Functional API #####
input_neurons = X_train.shape[1]
output_neurons = 1

x = Input(shape=(input_neurons))

# define hidden layers and neuron in each layer
number_of_hidden_layers = 2 # || layers
neuron_hidden_layer_1_1 = 10 # for first || layer
neuron_hidden_layer_1_2 = 20 # for second || layer

# both layer are in parallel as both are taking x as input which is from input layer
hidden1_1 = Dense(units=neuron_hidden_layer_1_1,activation='relu')(x)
hidden1_2 = Dense(units=neuron_hidden_layer_1_2,activation='relu')(x)
combined = concatenate([hidden1_1,hidden1_2]) # combining output of hidden1_1 and hidden1_2 to give it as an input for next layer

#second layer
neuron_hidden_layer_2 = 5
hidden2 = Dense(units=neuron_hidden_layer_2,activation='relu')(combined)
output = Dense(units=output_neurons,activation='sigmoid')(hidden2) # output layer


model_functional = Model(x,output) 
model_functional
```

### 1. `GrayScale images`
* Stored as 2-D matrix
* Matric Shape = Height * Width ==> Pixels across h and w
* Range of pixel values: 0 to 255
    * represent shade or intensity of pixels
    
### 2. `RGB images`
* Stored as 3-D matrix
* Matric Shape = Height * Width * Channel ==> Pixels across h and w
* `Number of Channels are 3` (idea is all other color can be made out of these three colors)
    * Red, Green, Blue
* Each channel are stacked together 
    * Three 2-D matric are stacked together
* Each individual matric have a range of 0 to 255 

### 3. `RGBA images`
* Stored as 3-D matrix
* Matric Shape = Height * Width * Channel ==> Pixels across h and w
* `Number of Channels are 4`
    * Red, Green, Blue, Alpha
* `Alpha` => controls Tranparency or Opacity of image
    * Range of `Alpha` - 0 to 255
        * `0 => occluded`
        * `255 => original image as it is`


### 4. `MRI images`
* Stored as 4-D matrix
* Matric Shape = Height * Width * Slices * Channel


<center> <h4 style="background-color:orange; color:white" ><br>Extracting Edges from Images<br></h4></center>

[Extract Features from Image Data using Python](https://www.analyticsvidhya.com/blog/2019/08/3-techniques-extract-features-from-image-data-machine-learning-python/) <br>
If from any image we will remove Color, Backgroud and other minute detailes we can still understand image only by observing Edges of image
* Removing color and background is call removing `Noise from images`

# Why we use edge detection?
* Reduce unnecessary information in image while preserving structure of image
* Extract important features of an image such as corners, lines, and curves
* Edges provide strong visual clues that can help recognition process

[Edge Detection Sumary:](https://www.bbau.ac.in/dept/CS/TM/Edge%20detection%20.pdf)

<center><img src='https://cv-tricks.com/wp-content/uploads/2019/03/imagematrix.png'></center>

Observe that Edges have different number then other parts

* Edges often represent boundary of objects present in an image
* Pixel values around Edges shows sudden change from left to right or from up to down 

### We can find which pixels lies on Edge:
1. To identify if a pixel is an edge or not, we will simply subtract values which are on either side of pixel `say 23|34|45 then sub 45 and 23 and see difference`
    * `If difference between pixels in either side of target pixel is larger` or above threshold => can conclude that there is a significant transition at this pixel and hence it is an edge
    * `If difference between pixels in either side of target pixel is smaller` or bellow threshold => can say that there is no edge around this pixel
    

### Kernels or Filters that can be used to highlight edges in an image
Above task of Edge detection can be achieved using `Prewitt Kernel` (in x-direction)
<center><img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2019/08/article-image-132.png'></center>

* We take values surrounding selected pixel and multiply it with `selected kernel(Prewitt kernel)` 
* We can then add resulting values to get a final value
* Since we already have `-1 in one column and 1 in the other column`, adding values is equivalent to taking difference
<center><img src='https://cdn.analyticsvidhya.com/wp-content/uploads/2019/08/article-image-111.png'></center>


## Other most popularly kernels or Filters
* `Prewitt Kernel X-Direction` => compares pixel values along `Horizontal axis`
* `Prewitt Kernel Y-Direction` => compares pixel values along `Vertical axis`
* `Sobel Kernel X-Direction` => in this kernel high importance is given to pixel values right next to target pixel
* `Sobel Kernel Y-Direction`

<center><img src='https://drive.google.com/file/d/1D-iQpqyQMoatLsVvtqZdtVpCqr5GWky6/view?usp=sharing'></center>


A Filter or Kernel(matric) slides through whole matrix of image,which produces a new matric which is called **`Feature Map`** <br>
Values of **`Feature Map`** denotes if a particular pixel lies on the edge or not

```python
from skimage.filters import prewitt_h,prewitt_v,sobel_h,sobel_v

#reading image 
image = imread(path+'dog.jpg',as_gray=True)


#calculating vertical edges using prewitt kernel
vertical_edges_y_direction = prewitt_v(image) # only vertical edge will not be highlighted

imshow(vertical_edges_y_direction,cmap='gray')
plt.show()
```

```python
#calculating horizontal edges using prewitt kernel
horizontal_edges_x_direction = prewitt_h(image) # only horizontal edge will not be highlighted

imshow(horizontal_edges_x_direction,cmap='gray')
plt.show()
```

### 1. Loading Data

```python
# import necessary libraries and functions
import numpy as np #here to convert images to array
import pandas as pd
from glob import glob #for reeding multiple images all together
import matplotlib.pyplot as plt
%matplotlib inline

from keras.layers import Dense, InputLayer
from keras.models import Sequential

from sklearn.model_selection import train_test_split

#create random number generator
seed = 108
rng = np.random.RandomState(seed)
```

```python
img_path = '/kaggle/input/emergency-vehicles-identification/Emergency_Vehicles/'
data = pd.read_csv(img_path+'train.csv')
```

```python
# any 10 samples
data.sample(10)
```

```python
# class distribution in %tg
data.emergency_or_not.value_counts(normalize=True)
```

Class-1 => Emergency Vehicle

```python
#loading images and storing them in for of numpy array i.e metric

#empty list to store images
X = []
#iterating over each image
for img_name in data.image_names:
    # loading image using its name
    test_img_path = '/kaggle/input/emergency-vehicles-identification/Emergency_Vehicles/train/'
    img = plt.imread(test_img_path + img_name)
    # saving each image in list
    X.append(img)
```

```python
#converting list of images into array and saving into X
X = np.array(X)

#storing target variable in separate variable
y = data.emergency_or_not.values
```

```python
X.shape
```

* `1646` => Number of Images
* `224, 224` => height and width of every image
* `3` => number of channels in each of these image i.e. images are in RGB format

```python
#visualizing some images
fig,ax = plt.subplots(nrows=1,ncols=5,figsize=(20,20))
for i in range(5):
    ax[i].imshow(X[i*400])
    ax[i].axis('off')
```

### 2. Pre-processing data

As NN takes only 1-D input but images are in 3-D, so changing them to 1_d

```python
# converting 3 dimensional image to 1 dimensional image
X = X.reshape(X.shape[0],224*224*3) #(number of images,new_shape of each image)
X.shape
```

After reshaping we have:
* `1646` => images
* `150528` => feature for each image


```python
# minimum and maximum pixel values of images
X.min(), X.max()
```

### normalizing pixel values
Bringing them down in range of 0 to 1

```python
X = X / X.max()
```

```python
# after normalizing, minimum and maximum pixel values of images 
X.min(), X.max()
```

0 => white pixel intesity <br>
1 => black pixel intensity

Will help to reduce training time and will use less computation

### 3. Creating training and validation set

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=seed)

(X_train.shape,y_train.shape),(X_test.shape,y_test.shape)
```

1152, 494 => number of images

### 4. Defining model architecture

```python
X_train.shape[1]
```

```python
model=Sequential()

model.add(InputLayer(input_shape=(X_train.shape[1]))) # 

model.add(Dense(100,activation='sigmoid'))
model.add(Dense(units=1,activation='sigmoid')) #binary classification problem so output neuron is 1

# summary of the model
model.summary()
```

### 5. Compiling Model

```python
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

### 6. Training model

```python
model_history = model.fit(X_train,y_train,epochs=50,batch_size=128,validation_data=(X_test,y_test))
```

`batch_size` => Number of training examples used in single Iteration <br>
`batch_size=128` => means 128 images will be loaded and image will be trained on these 128 images <br>
`model_history` => to save loss and accuracy values to compare later

### 7. Evaluating model performance

```python
# getting predictions in form of class as well as probabilities
#predictions = model.predict_classes(X_test)[:, 0] # class form of output

prediction_probabilities = model.predict(X_test)[:,0] #prob form of output
```

```python
# converting probabilities to classes
prediction_int = (prediction_probabilities >= 0.5).astype("int32")
```

```python
# accuracy on validation set
from sklearn.metrics import accuracy_score

print('Accuracy on validation set:', accuracy_score(y_test,prediction_int), '%')
```

```python
#summarizing history for loss

plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'], loc='upper right')
plt.show()
```

```python
#summarizing history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
# pull out the original images from data which correspond to the validation data
_, valid_vehicles, _, valid_y = train_test_split(data.image_names.values,y,test_size=0.3,random_state=seed) # using original data i.e. not preprocessed one

# get a random index to plot image randomly
index = rng.choice(range(len(valid_vehicles)))

# get the corresponding image name and probability
img_name = valid_vehicles[index]
prob = (prediction_probabilities * 100).astype(int)[index]

# read the image
img = plt.imread(test_img_path + img_name)

# print probability and actual class
print('Model is', prob , '% sure that this is emergency')
print('And actual class is ', valid_y[index])

# plot image
plt.imshow(img)
plt.show()
```
